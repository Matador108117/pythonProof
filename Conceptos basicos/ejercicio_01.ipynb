```python
import matplotlib.pyplot as plt
lista = [{"materia": "historia", "calf":90}, {"materia":"matematicas","calf":75 },
         {"materia": "ingles", "calf": 85}]
plt.bar([fila["materia"] for fila in lista], [int(fila["calf"]) for fila in lista], color="skyblue", edgecolor = "black" )
plt.title("Calificaciones por materia")
plt.xlabel("Materia")
plt.ylabel("Calificación")
```
