Ejemplos de codigo para numpy

```python
import numpy as np
# 1 crear vectores
lista = [1, 2, 3, 4, 5]
array = np.array(lista)
print(array)
# Salida: [1 2 3 4 5]

# 2 creacion de matrices
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz)

# salida [[1 2 3]
#        [4 5 6]]

# 3 operaciones basicas 

array = np.array([[1, 2, 3], [4, 5, 6]])

print("Matriz original:")
print(array)

# Suma de todos los elementos
print("\nSuma de todos los elementos:", np.sum(array))

# Transpuesta de la matriz
print("\nMatriz transpuesta:")
print(array.T)
# multiplicacion de matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = np.multiply(A, B)  # O simplemente A * B
print(result)
# Salida: [[ 5 12]
#          [21 32]]

# resta de matrices 
matriz1 = np.array([[1, 2, 3], [4, 5, 6]])
matriz2 = np.array([[6, 5, 4], [3, 2, 1]])

# Resta de matrices
resultado = matriz1 - matriz2

print("Resultado de la resta:")
print(resultado)

# Definimos las matrices
matriz1 = np.array([[1, 2, 3], [4, 5, 6]])
matriz2 = np.array([[6, 5, 4], [3, 2, 1]])

# Resta de matrices
resultado = np.subtract(matriz1, matriz2)

print("Resultado de la resta:")
print(resultado)


```

Ejemplo de codigo para Pandas

```python
import pandas as pd
# 1 abrir un csv
df = pd.read_csv('jugadores.csv')
df.head()
# 2 convertir de una estructura de datos a csv
datos = {
    "Nombre": ["Ana", "Luis", "Carlos", "María"],
    "Edad": [23, 34, 29, 31],
    "Ciudad": ["Orizaba", "Puebla", "Monterrey", "Guadalajara"]
}
df2 = pd.DataFrame(datos)
df2.to_csv("datos.csv")
df2.head()
# 3 modificar data frames 
data = {"Auto": ["Tacoma","Hilux","corolla"],
        "Kilometraje": ["10,000","80,000","1200"],
        "Color": ["Rojo","Negro","Gris"]
       }
df3 = pd.DataFrame(data)
# añadir una columna
df3.insert(3,"Velocidad",["160 km","149 km","90 km"])
# eliminar columna
df3 = df3.drop(columns=['Color'])
#añadir fila
df3.loc[len(df3)] = {'Auto':'camry','Kilometraje':'120,000','Velocidad':'120 km'} # la añade al final 
# eliminar fila
df3 = df3.drop(index=[0])
df3.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 1 grafico de dispersion
df = sns.load_dataset("penguins")  # dataset prehecho de pingüinos
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.show()
# 2 histograma
sns.histplot(data=df, x="flipper_length_mm", bins=20, kde=True)
plt.show()
#3 diagrama de cajas 
sns.boxplot(data=df, x="species", y="bill_length_mm")
plt.show()
```
